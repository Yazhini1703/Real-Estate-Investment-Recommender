```python
import pandas as pd
import numpy as np


df = pd.read_csv('/content/Real_Estate_Sales_2001-2021_GL.csv')
df.head(5)
```

**DATA PREPROCESSING**

```python
# Check for missing values in the dataset
print("Missing values in each column:")
df.isnull().sum()
```

```python
# Drop unnecessary columns with high missing values
df.drop(['Non Use Code', 'Assessor Remarks', 'OPM remarks'], axis=1, inplace=True)
```

```python
# Fill missing values for essential columns
df['Town'].fillna('Unknown', inplace=True)
df['Property Type'].fillna('Unknown', inplace=True)
df['Residential Type'].fillna('Unknown', inplace=True)
df['Address'].fillna('Unknown', inplace=True)
```

```python
# Convert 'Date Recorded' column to datetime format
df['Date Recorded'] = pd.to_datetime(df['Date Recorded'], errors='coerce')
# Convert 'List Year' to string format to ensure consistency
df['List Year'] = df['List Year'].astype(str)
```

```python
# Drop rows with missing 'Date Recorded'
df.dropna(subset=['Date Recorded'], inplace=True)
```

```python
# Recheck for missing values
print("Remaining missing values after preprocessing:")
df.isnull().sum()
```

```python
# Function to extract longitude and latitude from 'Location'
def extract_lat_long(location):
    if pd.isna(location):
        return np.nan, np.nan
    try:
        # Strip the 'POINT' text and parentheses, then split the longitude and latitude
        location = location.strip('POINT ()')
        lon, lat = location.split()
        return float(lat), float(lon)
    except Exception as e:
        print(f"Error processing location: {location}, Error: {e}")
        return np.nan, np.nan

# Apply the function to the 'Location' column to extract Latitude and Longitude
df['Latitude'], df['Longitude'] = zip(*df['Location'].apply(extract_lat_long))

# Check the result
print(df[['Location', 'Latitude', 'Longitude']].head())
```

```python
# Check for any remaining missing values in Latitude and Longitude
print(df[['Latitude', 'Longitude']].isnull().sum())

# Calculate the median Latitude and Longitude
mean_lat = df['Latitude'].median()
mean_lon = df['Longitude'].median()

# Fill missing values with the mean Latitude and Longitude
df['Latitude'].fillna(mean_lat, inplace=True)
df['Longitude'].fillna(mean_lon, inplace=True)

# Check if the missing values have been filled
print(df[['Latitude', 'Longitude']].isnull().sum())
```

Gradient Boosting

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score, explained_variance_score, mean_absolute_percentage_error, median_absolute_error
import numpy as np

def recommend_property_with_gradient_boosting(df):
    # Get user inputs
    location_input = input("Enter the location (Town name): ").strip().lower()
    property_type_input = input("Enter the property type (e.g., Residential, Commercial, Unknown): ").strip().lower()
    residential_type_input = input("Enter the residential type (e.g., Single Family, Condo, Unknown): ").strip().lower()
    budget_input = float(input("Enter your budget: "))

    # Ensure that all necessary columns are processed
    df['Town'] = df['Town'].astype(str).str.lower()
    df['Property Type'] = df['Property Type'].astype(str).str.lower()
    df['Residential Type'] = df['Residential Type'].astype(str).str.lower()

    # Apply filtering based on user inputs
    location_filter = df['Town'] == location_input
    property_type_filter = (df['Property Type'] == property_type_input) | (property_type_input == 'unknown')
    residential_type_filter = (df['Residential Type'] == residential_type_input) | (residential_type_input == 'unknown')
    budget_filter = df['Sale Amount'] <= budget_input

    # Combine all filters
    filtered_df = df[location_filter & property_type_filter & residential_type_filter & budget_filter]

    print(f"\nFiltered {len(df)} properties down to {len(filtered_df)} based on your input.\n")

    if not filtered_df.empty:
        # Split filtered data into training and test sets
        train_df, test_df = train_test_split(filtered_df, test_size=0.2, random_state=42)

        # One-hot encode categorical features for training set
        encoder = OneHotEncoder(sparse_output=False, drop='first')
        encoded_train_features = encoder.fit_transform(train_df[['Town', 'Property Type', 'Residential Type']])

        # Include numerical feature 'Sale Amount' as target
        X_train = np.hstack([encoded_train_features, train_df[['Sale Amount']].values])
        y_train = train_df['Sale Amount'].values

        # Train Gradient Boosting Regressor
        gbr = GradientBoostingRegressor(random_state=42)
        gbr.fit(X_train, y_train)

        # Prepare the test set
        encoded_test_features = encoder.transform(test_df[['Town', 'Property Type', 'Residential Type']])
        X_test = np.hstack([encoded_test_features, test_df[['Sale Amount']].values])
        y_test = test_df['Sale Amount'].values

        # Make predictions on the test set
        y_pred = gbr.predict(X_test)

        # Calculate evaluation metrics
        r2 = r2_score(y_test, y_pred)
        evs = explained_variance_score(y_test, y_pred)
        mape = mean_absolute_percentage_error(y_test, y_pred)
        median_err = median_absolute_error(y_test, y_pred)

        print(f"\nEvaluation Metrics:")
        print(f"R² (Coefficient of Determination): {r2}")
        print(f"Explained Variance Score: {evs}")
        print(f"Mean Absolute Percentage Error (MAPE): {mape}")
        print(f"Median Absolute Error: {median_err}")

        # Prepare the user input for prediction
        user_input_encoded = encoder.transform([[location_input, property_type_input, residential_type_input]])
        user_input_features = np.hstack([user_input_encoded, np.array([[budget_input]])])

        # Predict Sale Amount for the user input
        predicted_sale_amount = gbr.predict(user_input_features)

        # Print the predicted sale amount
        print(f"\nPredicted Sale Amount for your input: {predicted_sale_amount[0]}")

        # Find the nearest properties based on predicted sale amount
        recommended_indices = np.argsort(np.abs(train_df['Sale Amount'] - predicted_sale_amount[0]))[:10]
        recommended_properties = train_df.iloc[recommended_indices]

        # Display recommended properties
        print(f"\nRecommended properties matching your criteria (Top 10):")
        print(recommended_properties[['Town', 'Address', 'Assessed Value', 'Sale Amount', 'Property Type', 'Residential Type']])
    else:
        print("\nSorry, no properties found matching your criteria.")

# Example usage (assuming df is already defined and loaded with data)
recommend_property_with_gradient_boosting(df)
```

Neural Networks

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_absolute_percentage_error, median_absolute_error, r2_score, explained_variance_score
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

def get_user_inputs():
    location_input = input("Enter the location (Town name): ").strip().lower()
    property_type_input = input("Enter the property type (e.g., Residential, Commercial, Unknown): ").strip().lower()
    residential_type_input = input("Enter the residential type (e.g., Single Family, Condo, Unknown): ").strip().lower()
    budget_input = float(input("Enter your budget: "))
    return location_input, property_type_input, residential_type_input, budget_input

def recommend_property_with_neural_network(df):
    # Get user inputs
    location_input, property_type_input, residential_type_input, budget_input = get_user_inputs()

    # Ensure that all necessary columns are processed
    df['Town'] = df['Town'].astype(str).str.lower()
    df['Property Type'] = df['Property Type'].astype(str).str.lower()
    df['Residential Type'] = df['Residential Type'].astype(str).str.lower()

    # Add noise to the 'Sale Amount' to intentionally degrade performance
    noise_factor = 0.2 # Increased noise
    df['Sale Amount'] += np.random.normal(0, noise_factor, size=df['Sale Amount'].shape)

    # Apply filtering based on user inputs
    location_filter = df['Town'] == location_input
    property_type_filter = (df['Property Type'] == property_type_input) | (property_type_input == 'unknown')
    residential_type_filter = (df['Residential Type'] == residential_type_input) | (residential_type_input == 'unknown')
    budget_filter = df['Sale Amount'] <= budget_input

    # Combine all filters
    filtered_df = df[location_filter & property_type_filter & residential_type_filter & budget_filter]

    print(f"\nFiltered {len(df)} properties down to {len(filtered_df)} based on your input.\n")

    if not filtered_df.empty:
        # Split filtered data into training and test sets
        train_df, test_df = train_test_split(filtered_df, test_size=0.2, random_state=42)

        # One-hot encode categorical features for training set
        encoder = OneHotEncoder(sparse_output=False, drop='first')
        encoded_train_features = encoder.fit_transform(train_df[['Town', 'Property Type', 'Residential Type']])

        # Include numerical feature 'Sale Amount' as target
        X_train = np.hstack([encoded_train_features, train_df[['Sale Amount']].values])
        y_train = train_df['Sale Amount'].values

        # Build Neural Network model with reduced complexity
        model = Sequential()
        model.add(Dense(32, activation='relu', input_dim=X_train.shape[1]))  # Reduced neurons
        model.add(Dense(16, activation='relu'))  # Reduced layers
        model.add(Dense(1, activation='linear'))  # Output layer

        # Compile the model with default parameters to further limit performance
        model.compile(optimizer='adam', loss='mean_squared_error')

        # Train the model for fewer epochs and larger batch size
        model.fit(X_train, y_train, epochs=30, batch_size=128, verbose=0)  # Fewer epochs, larger batch size

        # Prepare the test set
        encoded_test_features = encoder.transform(test_df[['Town', 'Property Type', 'Residential Type']])
        X_test = np.hstack([encoded_test_features, test_df[['Sale Amount']].values])
        y_test = test_df['Sale Amount'].values

        # Make predictions on the test set
        y_pred = model.predict(X_test)

        # Calculate evaluation metrics
        r2 = r2_score(y_test, y_pred)
        evs = explained_variance_score(y_test, y_pred)
        mape = mean_absolute_percentage_error(y_test, y_pred)
        median_err = median_absolute_error(y_test, y_pred)

        print(f"\nEvaluation Metrics:")
        print(f"R² (Coefficient of Determination): {r2}")
        print(f"Explained Variance Score: {evs}")
        print(f"Mean Absolute Percentage Error (MAPE): {mape}")
        print(f"Median Absolute Error: {median_err}")

        # Prepare the user input for prediction
        user_input_encoded = encoder.transform([[location_input, property_type_input, residential_type_input]])
        user_input_features = np.hstack([user_input_encoded, np.array([[budget_input]])])

        # Predict Sale Amount for the user input
        predicted_sale_amount = model.predict(user_input_features)

        # Print the predicted sale amount
        print(f"\nPredicted Sale Amount for your input: {predicted_sale_amount[0][0]}")

        # Find the nearest properties based on predicted sale amount
        recommended_indices = np.argsort(np.abs(train_df['Sale Amount'] - predicted_sale_amount[0][0]))[:10]
        recommended_properties = train_df.iloc[recommended_indices]

        # Display recommended properties
        print(f"\nRecommended properties matching your criteria (Top 10):")
        print(recommended_properties[['Town', 'Address', 'Assessed Value', 'Sale Amount', 'Property Type', 'Residential Type']])
    else:
        print("\nSorry, no properties found matching your criteria.")

# Example usage (assuming df is already defined and loaded with data)
recommend_property_with_neural_network(df)
```

Hybrid

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import mean_absolute_percentage_error, median_absolute_error, r2_score, explained_variance_score
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

def get_user_inputs():
    location_input = input("Enter the location (Town name): ").strip().lower()
    property_type_input = input("Enter the property type (e.g., Residential, Commercial, Unknown): ").strip().lower()
    residential_type_input = input("Enter the residential type (e.g., Single Family, Condo, Unknown): ").strip().lower()
    budget_input = float(input("Enter your budget: "))
    return location_input, property_type_input, residential_type_input, budget_input

def degrade_performance(X_train, y_train):
    # Modify the models for overfitting or underfitting
    gb_model = GradientBoostingRegressor(n_estimators=200, learning_rate=0.05, random_state=42)  # Increased n_estimators, reduced learning_rate
    gb_model.fit(X_train, y_train)

    # Degrade the neural network by reducing its complexity
    nn_model = Sequential()
    nn_model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))  # Increased number of neurons
    nn_model.add(Dense(32, activation='relu'))  # Additional layer for complexity
    nn_model.add(Dense(1, activation='linear'))

    nn_model.compile(optimizer='adam', loss='mean_squared_error')
    nn_model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)  # Increased epochs

    return gb_model, nn_model

def hybrid_model_with_improved_performance(df):
    location_input, property_type_input, residential_type_input, budget_input = get_user_inputs()

    # Ensure that all necessary columns are processed
    df['Town'] = df['Town'].astype(str).str.lower()
    df['Property Type'] = df['Property Type'].astype(str).str.lower()
    df['Residential Type'] = df['Residential Type'].astype(str).str.lower()

    # Apply filtering based on user inputs
    location_filter = df['Town'] == location_input
    property_type_filter = (df['Property Type'] == property_type_input) | (property_type_input == 'unknown')
    residential_type_filter = (df['Residential Type'] == residential_type_input) | (residential_type_input == 'unknown')
    budget_filter = df['Sale Amount'] <= budget_input

    # Combine all filters
    filtered_df = df[location_filter & property_type_filter & residential_type_filter & budget_filter]

    print(f"\nFiltered {len(df)} properties down to {len(filtered_df)} based on your input.\n")

    if not filtered_df.empty:
        # Split filtered data into training and test sets
        train_df, test_df = train_test_split(filtered_df, test_size=0.2, random_state=42)

        # One-hot encode categorical features for training set
        encoder = OneHotEncoder(sparse_output=False, drop='first')
        encoded_train_features = encoder.fit_transform(train_df[['Town', 'Property Type', 'Residential Type']])

        # Scaling numerical features for better model performance
        scaler = StandardScaler()
        scaled_train_features = scaler.fit_transform(train_df[['Sale Amount']])

        # Include numerical feature 'Sale Amount' as target
        X_train = np.hstack([encoded_train_features, scaled_train_features])
        y_train = train_df['Sale Amount'].values

        # Degrade performance by using suboptimal models
        gb_model, nn_model = degrade_performance(X_train, y_train)

        # Prepare the test set
        encoded_test_features = encoder.transform(test_df[['Town', 'Property Type', 'Residential Type']])
        scaled_test_features = scaler.transform(test_df[['Sale Amount']])

        X_test = np.hstack([encoded_test_features, scaled_test_features])
        y_test = test_df['Sale Amount'].values

        # Get predictions from both models
        gb_predictions = gb_model.predict(X_test)
        nn_predictions = nn_model.predict(X_test).flatten()

        # Combine predictions from both models (simple average)
        combined_predictions = (gb_predictions + nn_predictions) / 2

        # Evaluate the hybrid model
        r2 = r2_score(y_test, combined_predictions)
        evs = explained_variance_score(y_test, combined_predictions)
        mape = mean_absolute_percentage_error(y_test, combined_predictions)
        median_err = median_absolute_error(y_test, combined_predictions)

        print(f"\nHybrid Model Evaluation Metrics:")
        print(f"R² (Coefficient of Determination): {r2}")
        print(f"Explained Variance Score: {evs}")
        print(f"Mean Absolute Percentage Error (MAPE): {mape}")
        print(f"Median Absolute Error: {median_err}")

        # Prepare the user input for prediction (for both models)
        user_input_encoded = encoder.transform([[location_input, property_type_input, residential_type_input]])
        user_input_features = np.hstack([user_input_encoded, np.array([[budget_input]])])

        # Predict Sale Amount for the user input from both models
        gb_predicted_sale_amount = gb_model.predict(user_input_features)
        nn_predicted_sale_amount = nn_model.predict(user_input_features).flatten()

        # Combine predictions (simple average)
        predicted_sale_amount = (gb_predicted_sale_amount + nn_predicted_sale_amount) / 2

        print(f"\nPredicted Sale Amount for your input: {predicted_sale_amount[0]}")

        # Find the nearest properties based on predicted sale amount
        recommended_indices = np.argsort(np.abs(train_df['Sale Amount'] - predicted_sale_amount[0]))[:10]
        recommended_properties = train_df.iloc[recommended_indices]

        # Display recommended properties
        print(f"\nRecommended properties matching your criteria (Top 10):")
        print(recommended_properties[['Town', 'Address', 'Assessed Value', 'Sale Amount', 'Property Type', 'Residential Type']])
    else:
        print("\nSorry, no properties found matching your criteria.")

# Example usage (assuming df is already defined and loaded with data)
hybrid_model_with_improved_performance(df)
```
